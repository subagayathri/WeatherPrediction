```python
#Reqiured modules
import pandas as pd
from sklearn.preprocessing import  LabelEncoder
from sklearn.naive_bayes import  GaussianNB
```

```python
#Reading the csv file
df=pd.read_csv("weather.csv")
df
```

```python
# reading the csv files
df=pd.read_csv("weather.csv")
df
#Encoding the strings to Numericals

Numerics = LabelEncoder()

#Processing the target variable and make it is as newframe
inputs=df.drop('play', axis='columns')
target=df['play']
target
```

```python
#Creating the new DataFrame


inputs['Outlook_n']=Numerics.fit_transform(inputs['outlook'])
inputs['Temp_n']=Numerics.fit_transform(inputs['temperature'])
inputs['Hum_n']=Numerics.fit_transform(inputs['humidity'])
inputs['Win_n']=Numerics.fit_transform(inputs['windy'])
inputs
```

```python
# dropping the string values
inputs_n=inputs.drop(['outlook','temperature','humidity','windy'],axis='columns')
inputs_n
```

```python
#Applying Gaussian naivebayes
classifier=GaussianNB()
classifier.fit(inputs_n,target)

GaussianNB()

#92% accuracy
classifier.score(inputs_n,target)
```

```python
#Prediction
classifier.predict([[1,0,0,1]])
```
